# Download The data

```python
import requests
import zipfile
from pathlib import Path

# Setup path to data folder
data_path = Path("data/")
image_path = data_path / "pizza_steak_sushi"

# Check if the directory exists 
if image_path.is_dir():
    print(f"{image_path} directory already exists, skipping download and extraction.")
else:
    print(f"No direcotory found at {image_path}, creating one and downloading data.")
    image_path.mkdir(parents=True, exist_ok=True)

    # Download the data
    with open(image_path / "pizza_steak_sushi.zip", "wb") as f:
        request = requests.get("https://github.com/mrdbourke/pytorch-deep-learning/raw/main/data/pizza_steak_sushi.zip")
        print("Downloading data...")
        f.write(request.content)

    # Unzip the data
    with zipfile.ZipFile(image_path / "pizza_steak_sushi.zip", "r") as zip_ref:
        print("Extracting data...")
        zip_ref.extractall(image_path)
```

# Data Preparation

```python
import os 

def walk_through_dir(dir_path):
    """
    Walks through dir_path returning its contents.
    Args:
        dir_path: a target directory path as a string or pathlib.Path
    Returns:
        A print out of the directory structure.
    """
    for dirpath, dirnames, filenames in os.walk(dir_path):
        print(f"There are {len(dirpath)} directories and {len(dirnames)} files and images {len(filenames)} in '{dirpath}'")
```

```python
walk_through_dir(image_path)
```
